```
def accuracy(predictions,labels):
    if (len(predictions) != len(labels)):
        print("they are not equel")
        return 0
    counter=0
    for i in range(len(labels)):
        if (predictions[i] == labels[i]):
            counter+=1
    return counter/len(labels)
```

```
def print_model_results( model_name, model, datasets, beta=2, cv=KFold(n_splits=3), show_out=True):
  """
  name of model, model, (trainx,trainy), beta score, crossvalidation, print report if true returns vals if false
  """
  base_model = make_pipeline(StandardScaler(),model)

  num_of_splits = cv.get_n_splits()

  scores = []
  fold_ind = 0
  for train_ind, tune_ind in cv.split(datasets['train_x']):
    clf = sk.base.clone(base_model)
    train_x = datasets['train_x'].iloc[train_ind]
    train_y = datasets['train_y'].iloc[train_ind]
    clf.fit( train_x, train_y )

    tuning_x = datasets['train_x'].iloc[tune_ind]
    tuning_y = datasets['train_y'].iloc[tune_ind]

    tuning_data_final_pred_results = clf.predict( tuning_x )
    confusion = confusion_matrix(tuning_y, tuning_data_final_pred_results)
    tn, fp, fn, tp = confusion.ravel()
    scores.append({
        'model' : clf,
        'fold': fold_ind,
        'accuracy': accuracy_score(tuning_y, tuning_data_final_pred_results),
        'fbeta': fbeta_score(tuning_y, tuning_data_final_pred_results, average='binary', beta=beta),
        'tn': tn,
        'fp': fp,
        'fn': fn,
        'tp': tp,
        'confusion': confusion
    })
    fold_ind += 1
  
  acc_list = [ x['accuracy'] for x in scores ]
  avg_acc = sum( acc_list ) / num_of_splits

  fbeta_list = [ x['fbeta'] for x in scores ]
  avg_fbeta = sum( fbeta_list ) / num_of_splits

  tp_list = [ x['tp'] for x in scores ]
  avg_tp = str( round( (sum( tp_list ) / num_of_splits), 2) ).ljust(5)

  fp_list = [ x['fp'] for x in scores ]
  avg_fp = str( round( (sum( fp_list ) / num_of_splits), 2) ).ljust(5)

  fn_list = [ x['fn'] for x in scores ]
  avg_fn = str( round( (sum( fn_list ) / num_of_splits), 2) ).ljust(5)

  tn_list = [ x['tn'] for x in scores ]
  avg_tn = str( round( (sum( tn_list ) / num_of_splits), 2) ).ljust(5)
  
  base_model.fit(datasets['train_x'], datasets['train_y'])

  if( show_out ):
    print( (model_name + " Model Results").ljust(75, '-'))
    print( f'Average Accuracy Across {num_of_splits} splits'.ljust(33),':', f'{round(avg_acc,3)} | {pretty_format_list(acc_list)}' )
    print( f'Average fbeta Across {num_of_splits} splits'.ljust(33),':', f'{round(avg_fbeta,3)} | {pretty_format_list(fbeta_list)}' )
    print( f'Average tp Across {num_of_splits} splits'.ljust(33),':', avg_tp, f'| {pretty_format_list(tp_list)}' )
    print( f'Average fp Across {num_of_splits} splits'.ljust(33),':', avg_fp, f'| {pretty_format_list(fp_list)}' )
    print( f'Average fn Across {num_of_splits} splits'.ljust(33),':', avg_fn, f'| {pretty_format_list(fn_list)}' )
    print( f'Average tn Across {num_of_splits} splits'.ljust(33),':', avg_tn, f'| {pretty_format_list(tn_list)}' )
    print('Confusion Matrix'.ljust(75, '-'))
    for s in scores:
      print('[', str(s['tp']).ljust(2), ',',  str(s['fp']).ljust(2), ']', end=' ')

    print()
    for s in scores:
      print('[', str(s['fn']).ljust(2), ',',  str(s['tn']).ljust(2), ']', end=' ')
    
    print(end='\n\n')

  return {
      'avg_acc': float(avg_acc),
      'avg_fbeta': float(avg_fbeta),
      'avg_tp': float(avg_tp),
      'avg_fp': float(avg_fp),
      'avg_fn': float(avg_fn),
      'avg_tn': float(avg_tn),
      'trials': scores,
      'base_model': base_model
  }
```

```
def pretty_format_list( list_passed, round_to=2, max_len=4 ):
  list_to_work_with = [ str(round(x, round_to)) for x in list_passed]
  largest_size = max( [len(x) for x in list_to_work_with] ) if not max_len else max_len
  return [x.rjust(largest_size,' ') for x in list_to_work_with]
```

```
def find_value_to_remove( data_package, feature_set_given ):
  base_value = print_model_results( 
    'Linear SVC', 
    LinearSVC( random_state=random_state_for_model ),
    data_package,
    beta_for_eval,
    show_out=False
  )['avg_fbeta']
  # print(f'value to beat: {base_value}')

  scores_without_features = []
  for feature in feature_set_given:

    # feature set without current feature
    features_in_use = feature_set_given[:]
    features_in_use.remove(feature)

    data_package_without_feature = {
        'train_x': data_package['train_x'].drop( columns=[feature] ),
        'train_y': data_package['train_y']
    }

    # get fbeta for smaller feature set
    fbeta_without_feature = print_model_results( 
      'Linear SVC', 
      LinearSVC( random_state=random_state_for_model ),
      data_package_without_feature,
      beta_for_eval,
      show_out=False
    )['avg_fbeta']

    performed_better = fbeta_without_feature >= base_value

    if( performed_better ):
      scores_without_features.append( (feature, fbeta_without_feature, performed_better) )

    # print( f"Performace without {feature}:".ljust(40), f"{fbeta_without_feature} : {performed_better}" )

  if( len(scores_without_features) == 0):
    return None, base_value

  feature_to_remove = pd.DataFrame(
      data=scores_without_features, columns=['feature','score', 'performed_better_without']
  ).sort_values( by='score', ascending=False).iloc[0]

  return feature_to_remove['feature'], feature_to_remove['score']
```

```
def find_value_to_add( full_data_package, feature_set_given, current_feature_set ):

  current_data_package = {
        'train_x': full_data_package['train_x'][ current_feature_set ] ,
        'train_y': full_data_package['train_y']
    }

  base_value = print_model_results( 
    'Linear SVC', 
    LinearSVC( random_state=random_state_for_model ),
    current_data_package,
    beta_for_eval,
    show_out=False
  )['avg_fbeta']
  print(f'value to beat: {base_value}')

  scores_with_features = []
  for feature in feature_set_given:

    # print( f'testing: { [feature] + current_feature_set }' )
    new_data_package = {
        'train_x': full_data_package['train_x'][ current_feature_set + [feature] ] ,
        'train_y': full_data_package['train_y']
    }

    # get fbeta for smaller feature set
    fbeta_with_feature = print_model_results( 
      'Linear SVC', 
      LinearSVC( random_state=random_state_for_model ),
      new_data_package,
      beta_for_eval,
      show_out=False
    )['avg_fbeta']

    performed_better = fbeta_with_feature > base_value

    if( performed_better ):
      scores_with_features.append( (feature, fbeta_with_feature, performed_better) )

    # print( f"Performace with {feature}:".ljust(40), f"{fbeta_with_feature} : {performed_better}" )

  if( len(scores_with_features) == 0):
    return None, base_value

  feature_to_add = pd.DataFrame(
      data=scores_with_features, columns=['feature','score', 'performed_better_with']
  ).sort_values( by='score', ascending=False).iloc[0]

  return feature_to_add['feature'], feature_to_add['score']
```

```
#FROM AUDIOSEGMENTATION
def train_hmm_from_directory(folder_path, hmm_model_name, mid_window, mid_step):
    """
    This function trains a HMM model for segmentation-classification using
    a where WAV files and .segment (ground-truth files) are stored
    ARGUMENTS:
     - folder_path:     the path of the data diretory
     - hmm_model_name:  the name of the HMM model to be stored
     - mt_win:          mid-term window size
     - mt_step:         mid-term window step
    RETURNS:
     - hmm:            an object to the resulting HMM
     - class_names:    a list of class_names

    After training, hmm, class_names, along with the mt_win
    and mt_step values are stored in the hmm_model_name file
    """
    f_all = 0
    flags_all = np.array([])
    class_names_all = []
    for i, f in enumerate(glob.glob(folder_path + os.sep + '*.wav')):
        # for each WAV file
        wav_file = f
        gt_file = f.replace('.wav', '.segments')
        if os.path.isfile(gt_file):
            seg_start, seg_end, seg_labs = read_segmentation_gt(gt_file)
            flags, class_names = \
                segments_to_labels(seg_start, seg_end, seg_labs, mid_step)
            for c in class_names:
                # update class names:
                if c not in class_names_all:
                    class_names_all.append(c)
            sampling_rate, signal = audioBasicIO.read_audio_file(wav_file)
            feature_vector, _, _ = \
                mtf.mid_feature_extraction(signal, sampling_rate,
                                           mid_window * sampling_rate,
                                           mid_step * sampling_rate,
                                           round(sampling_rate * 0.050),
                                           round(sampling_rate * 0.050))

            flag_len = len(flags)
            feat_cols = feature_vector.shape[1]
            min_sm = min(feat_cols, flag_len)
            feature_vector = feature_vector[:, 0:min_sm]
            flags = flags[0:min_sm]

            flags_new = []
            # append features and labels
            for j, fl in enumerate(flags):
                flags_new.append(class_names_all.index(class_names_all[flags[j]]))

            flags_all = np.append(flags_all, np.array(flags_new))

            if i == 0:
                f_all = feature_vector
            else:
                f_all = np.concatenate((f_all, feature_vector), axis=1)

    # compute HMM statistics
    class_priors, transmutation_matrix, means, cov = \
        train_hmm_compute_statistics(f_all, flags_all)
    # train the HMM
    hmm = hmmlearn.hmm.GaussianHMM(class_priors.shape[0], "diag")
    hmm.covars_ = cov
    hmm.means_ = means
    hmm.startprob_ = class_priors
    hmm.transmat_ = transmutation_matrix

    save_hmm(hmm_model_name, hmm, class_names_all, mid_window, mid_step)

    return hmm, class_names_all
```

```

import os
from pydub import AudioSegment
from pydub.silence import split_on_silence
import speech_recognition as sr

# create a speech recognition object
r = sr.Recognizer()

# a function that splits the audio file into chunks
# and applies speech recognition
path='/content/drive/MyDrive/CSCE 5215/Spring 2021/Shortened Enki Audio Files/1CCC 2-19-18.wav'



def get_large_audio_transcription(path):
    """
    Splitting the large audio file into chunks
    and apply speech recognition on each of these chunks
    """
    # open the audio file using pydub
    sound = AudioSegment.from_wav(path)
    # split audio sound where silence is 700 miliseconds or more and get chunks
    chunks = split_on_silence(sound,
        # experiment with this value for your target audio file
        min_silence_len = 900,
        # adjust this per requirement
        silence_thresh = sound.dBFS - 14,
        # keep the silence for 1 second, adjustable as well
        keep_silence = 900,
)
    folder_name = "audio-chunks"
    # create a directory to store the audio chunks
    if not os.path.isdir(folder_name):
        os.mkdir(folder_name)
    whole_text = ""
    # process each chunk
    for i, audio_chunk in enumerate(chunks, start=1):
        # export audio chunk and save it in
        # the `folder_name` directory.
        chunk_filename = os.path.join(folder_name, f"chunk{i}.wav")
        audio_chunk.export(chunk_filename, format="wav")
        # recognize the chunk
        with sr.AudioFile(chunk_filename) as source:
            audio_listened = r.record(source)
            # try converting it to text
            try:
                text = r.recognize_google(audio_listened)
            except sr.UnknownValueError as e:
                print("Error:", str(e))
            else:
                text = f"{text.capitalize()}. "
                print(chunk_filename, ":", text)
                whole_text += text
    # return the text for all chunks detected
    return whole_text

get_large_audio_transcription(path)
```
